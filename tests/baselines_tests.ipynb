```python
import baselines

class TestNaiveModel(TestCase):
  # question why that works without executing?
    def setUp(self):
      self.dummy_series = pd.Series(np.array([40, 20]*100), index = pd.date_range(start = "2020-01-01", periods = 200, freq ='W-Mon'))
      self.model = baselines.NaiveForecast(horizon = 3)
      self.prediction = self.model.fit(self.dummy_series[:-3]).predict(self.dummy_series[-3:])

    def test_that_model_fit_predict_returns_correct(self):
      assert self.model.prediction_ == 40
      assert len(self.prediction) == 3 # same as horizon
      assert self.model.last_time_obs_ == pd.Timestamp('2023-10-09 00:00:00')
    def test_that_error_returns_correct(self):
      mse = mean_squared_error(self.dummy_series[-3:], self.prediction, squared=True)
      assert round(mse,4) == 266.6667
    def test_that_cross_validation_returns_correct(self):
      tscv = TimeSeriesSplit(n_splits=5, test_size = 3)
      cv_scores = cross_val_score(self.model, self.dummy_series, self.dummy_series,  cv=tscv, 
                            scoring='neg_mean_squared_error', n_jobs=1)
      print(cv_scores)
```
