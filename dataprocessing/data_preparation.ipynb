```python
def parse_text(code):
    #usunelam 'SMG_': code[3:6] bo wg dokumentacji zawsze ta sama wartosc
    szczegoly = code[17:28].replace("_", '')
    return {'oddzial': code[0:2], 'dane_alarmy': code[2], 'lok_funk_SAP':code[7:16], 'kopalnia':code[16], 'szczegoly':szczegoly, 'typ_zmiennej':code[28], 'jednostki_inzynierskie':code[29:]}
```

```python
def extract_time_from_path(path):
    result = re.search("[0-9]{2}-[0-9]{2}", path).group(0)
    time_detailed = dict(zip(["file_start_hour", "file_end_hour"], re.findall("[0-9]{1,}", result)))
    return time_detailed
```
